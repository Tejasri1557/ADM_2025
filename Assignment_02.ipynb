```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df2 = pd.DataFrame({'Age': np.random.randint(20, 60, size=100)})
plt.hist(df2['Age'], bins=30, color='blue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
```
