Dimensionality Reduction

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

# Load the dataset

```python
df = pd.read_csv('/content/Iris week 3.csv')
```

# Display first few rows to understand structure

```python
print(df.head())
```

# Assume the last column is the target and the rest are features

```python
target_column = df.columns[-1]
X = df.drop(columns=[target_column])
y = df[target_column]
```

# Standardize the data

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Apply PCA to reduce dimensions to 2

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

# Convert to DataFrame for visualization

```python
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df[target_column] = y
```

# Plot the transformed data

```python
plt.figure(figsize=(8, 6))
for label in np.unique(y):
    plt.scatter(pca_df[pca_df[target_column] == label]['PC1'],
                pca_df[pca_df[target_column] == label]['PC2'],
                label=label)

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Visualization of Reduced Features')
plt.legend()
plt.show()
```
